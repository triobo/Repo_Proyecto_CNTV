```python
print(df.columns)
```

```python
print(df['TEMA DENUNICIADO'].unique())
print(df['Canal / Cable Operador'].unique())
```

```python
resumen = df.groupby(['TEMA DENUNICIADO', 'Canal / Cable Operador']).size().reset_index(name='Cantidad')
resumen.head()
```

```python
import altair as alt

grafico = alt.Chart(resumen).mark_bar().encode(
    x=alt.X('Cantidad:Q', title='Cantidad de Sanciones'),  # Cantidad de sanciones, eje horizontal
    y=alt.Y('TEMA DENUNICIADO', title='Tema de la Sanción', sort='-x'),  # Tema en eje vertical, ordenado por cantidad descendente
    color='Canal / Cable Operador',  # El color de la barra indica el canal
    tooltip=['TEMA DENUNICIADO', 'Canal / Cable Operador', 'Cantidad']  # Información extra al pasar mouse
).properties(
    title='Cantidad de sanciones por tema y canal',
    width=700,
    height=400
)

grafico
```
