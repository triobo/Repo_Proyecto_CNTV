```python
# PASO 1: Subir archivo
from google.colab import files
import pandas as pd
import altair as alt

uploaded = files.upload()

# PASO 2: Leer archivo (detecta el nombre subido)
nombre_archivo = list(uploaded.keys())[0]
df = pd.read_excel(nombre_archivo)

# PASO 3: Revisar nombres originales de columnas
print("Columnas originales:", df.columns.tolist())

# PASO 4: Limpiar nombres (eliminar espacios antes y después)
df.columns = df.columns.str.strip()

# PASO 5: Renombrar columnas 0, 1 y 2 a nombres que quieres
df.columns.values[0:3] = ['AÑO', 'CANAL', 'MULTA TOTAL']

print("Columnas renombradas:", df.columns.tolist())

# PASO 6: Convertir AÑO a texto para Altair
df['AÑO'] = df['AÑO'].astype(str)

# PASO 7: Crear visualización Altair (barras apiladas)
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('AÑO:O', title='Año'),
    y=alt.Y('sum(MULTA TOTAL):Q', title='Multa Total (CLP)'),
    color=alt.Color('CANAL:N', title='Canal'),
    tooltip=['AÑO', 'CANAL', 'MULTA TOTAL']
).properties(
    width=700,
    height=400,
    title='Multas CNTV por canal y año'
)

chart.display()
```
