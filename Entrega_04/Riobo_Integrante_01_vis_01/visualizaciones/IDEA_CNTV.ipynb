```python
import pandas as pd
import altair as alt
from google.colab import files

# Assuming you've already uploaded the file and loaded the DataFrame df

# Paso 3: Ver todas las columnas del archivo para identificar la del programa
# Keep this step to verify column names
for col in df.columns:
    print(col)

# Convert 'Nº Denuncias' to numeric, coercing errors to NaN
df['Nº Denuncias'] = pd.to_numeric(df['Nº Denuncias'], errors='coerce')

# Drop rows where 'Nº Denuncias' became NaN after coercion (optional, depending on how you want to handle non-numeric entries)
df.dropna(subset=['Nº Denuncias'], inplace=True)


# Agrupar por tema y programa, sumando las denuncias
denuncias_por_programa = df.groupby(['Contenidos Denunciados', 'Programa'])['Nº Denuncias'].sum().reset_index()

# Sumar total de denuncias por programa
total_por_programa = denuncias_por_programa.groupby('Programa')['Nº Denuncias'].sum().reset_index()

# Obtener los 20 programas con más denuncias
top_20_programas = total_por_programa.sort_values('Nº Denuncias', ascending=False).head(20)['Programa']

# Filtrar el DataFrame original solo con esos 20 programas
denuncias_top20 = denuncias_por_programa[denuncias_por_programa['Programa'].isin(top_20_programas)]

# Crear el gráfico
chart_top20 = alt.Chart(denuncias_top20).mark_bar().encode(
    x=alt.X('Programa:N', title='Programa', sort='-y'),
    y=alt.Y('Nº Denuncias:Q', title='N° de Denuncias'),
    color=alt.Color('Contenidos Denunciados:N', title='Tema'),
    tooltip=['Contenidos Denunciados', 'Programa', 'Nº Denuncias']
).properties(
    width=800,
    height=500,
    title='Top 20 programas más denunciados por tema (2012–2022)'
).configure_axis(
    labelAngle=-45
)

chart_top20
```

```python
import pandas as pd
import altair as alt

# Convertir 'Nº Denuncias' a numérico (si no está hecho)
df['Nº Denuncias'] = pd.to_numeric(df['Nº Denuncias'], errors='coerce')
df.dropna(subset=['Nº Denuncias'], inplace=True)

# Diccionario para mapear los temas originales a las 5 categorías
categoria_temas = {
    # Completa o ajusta según tus temas reales en 'Contenidos Denunciados'
    'Violencia física': 'Violencia, Agresión y Conductas Inapropiadas',
    'Acoso sexual': 'Violencia, Agresión y Conductas Inapropiadas',
    'Lenguaje vulgar': 'Contenido Sexual / Lenguaje Vulgar',
    'Contenido sexual explícito': 'Contenido Sexual / Lenguaje Vulgar',
    'Difamación': 'Dignidad / Intimidad / Vida Privada',
    'Violación a la privacidad': 'Dignidad / Intimidad / Vida Privada',
    'Desinformación': 'Desinformación / Sesgo / Falta de Pluralismo / Manipulación',
    'Sesgo': 'Desinformación / Sesgo / Falta de Pluralismo / Manipulación',
    'Manipulación': 'Desinformación / Sesgo / Falta de Pluralismo / Manipulación',
    'Protección a la niñez': 'Niñez / Horario de Protección',
    'Contenido inadecuado para niños': 'Niñez / Horario de Protección',
    # Agrega los demás temas que tengas
}

# Mapear 'Contenidos Denunciados' a la categoría
df['Categoría'] = df['Contenidos Denunciados'].map(categoria_temas)

# Eliminar filas sin categoría asignada
df_filtrado = df.dropna(subset=['Categoría'])

# Agrupar por Categoría y Programa, sumando denuncias
denuncias_por_categoria_programa = df_filtrado.groupby(['Categoría', 'Programa'])['Nº Denuncias'].sum().reset_index()

# Sumar total de denuncias por programa para obtener top 20
total_por_programa = denuncias_por_categoria_programa.groupby('Programa')['Nº Denuncias'].sum().reset_index()

top_20_programas = total_por_programa.sort_values('Nº Denuncias', ascending=False).head(20)['Programa']

# Filtrar solo esos programas
denuncias_top20 = denuncias_por_categoria_programa[denuncias_por_categoria_programa['Programa'].isin(top_20_programas)]

# Orden para Altair
orden_programas = top_20_programas.tolist()

# Crear gráfico apilado por categoría y programa
chart_top20 = alt.Chart(denuncias_top20).mark_bar().encode(
    x=alt.X('Programa:N', title='Programa', sort=orden_programas),
    y=alt.Y('Nº Denuncias:Q', title='N° de Denuncias'),
    color=alt.Color('Categoría:N', title='Categoría'),
    tooltip=['Categoría', 'Programa', 'Nº Denuncias']
).properties(
    width=900,
    height=500,
    title='Top 20 programas más denunciados por categoría (2012–2022)'
).configure_axis(
    labelAngle=-45
)

chart_top20
```
