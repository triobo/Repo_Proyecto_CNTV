```python
import pandas as pd

# Leer el archivo subido (ajusta el nombre si es distinto)
df = pd.read_excel('/content/RANKING DE DENUNCIAS 2012-2022_FINAL1.xlsx')
df.head()
```

```python
# Primero revisa cómo se llama la columna que describe el tema de la denuncia
print(df.columns)
```

```python
# Función para agrupar los temas en las 5 categorías
def clasificar_tematica(texto):
    texto = str(texto).lower()
    if "violencia" in texto or "agresión" in texto or "inapropiado" in texto or "conducta" in texto:
        return "Violencia, Agresión y Conductas Inapropiadas"
    elif "dignidad" in texto or "intimidad" in texto or "privacidad" in texto or "vida privada" in texto:
        return "Dignidad / Intimidad / Vida Privada"
    elif "sexual" in texto or "vulgar" in texto or "lenguaje" in texto:
        return "Contenido Sexual / Lenguaje Vulgar"
    elif "desinformación" in texto or "sesgo" in texto or "pluralismo" in texto or "manipulación" in texto:
        return "Desinformación / Sesgo / Falta de Pluralismo / Manipulación"
    elif "niñez" in texto or "protección" in texto or "horario" in texto:
        return "Niñez / Horario de Protección"
    else:
        return "Otro"

# Aplica la función a la columna "Contenidos Denunciados"
df['Temática Agrupada'] = df['Contenidos Denunciados'].apply(clasificar_tematica)

# Muestra algunas filas para revisar
df[['Contenidos Denunciados', 'Temática Agrupada']].head(10)
```

```python
# Muestra todas las filas con la temática clasificada
df[['Contenidos Denunciados', 'Temática Agrupada']]
```

```python
agrupado = df.groupby(['Temática Agrupada', 'CANAL / CABLE ESTANDARIZADO']).size().reset_index(name='N° DENUNCIAS')
```

```python
import altair as alt

chart = alt.Chart(agrupado).mark_bar().encode(
    x=alt.X('CANAL / CABLE ESTANDARIZADO:N', title='Canal de Televisión'),
    y=alt.Y('N° DENUNCIAS:Q', title='N° DENUNCIAS'),
    color=alt.Color('Temática Agrupada:N', title='Temática'),
    tooltip=['Temática Agrupada', 'CANAL / CABLE ESTANDARIZADO', 'N° DENUNCIAS']
).properties(
    width=700,
    height=400,
    title='Denuncias por Temática y Canal (2012–2022)'
).configure_axis(
    labelAngle=-45
)

chart
```
